```python
import cv2
import numpy as np
import mediapipe as mp
from mediapipe.tasks.python import vision
from mediapipe.tasks import python
```

```python
path="face_landmarker.task"
options=python.BaseOptions(model_asset_path=path)
options_=vision.FaceLandmarkerOptions(options,num_faces=1)
```

```python
mesh_face=vision.FaceLandmarker.create_from_options(options_)
```

```python
img=cv2.imread(r"download.jpg")
```

```python
mp_img=mp.Image.create_from_file(r"download.jpg")
result=mesh_face.detect(mp_img)
```

```python
if result.face_landmarks:
    for land in result.face_landmarks:
        for lan_ in land:

                h, w, _ = img.shape
                x = int(lan_.x * w)
                y = int(lan_.y * h)
                cv2.circle(img, (x, y), 2, (0, 255, 0), -1)
cv2.imshow("FaceMesh", img)
path_save=r"C:\Users\abdal\OneDrive\Desktop\nti\mesh_face.jpg"

cv2.imwrite(path_save,img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
